```python
# 打开文件，替换 'your_file.txt' 为文件的实际路径
file_path = 'data3/train2.jsonl'
import json
codeTokenNum = []
nlTokenNum = []
try:
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()  # 读取文件内容，并将每行存储为列表中的元素
        line_count = len(lines)    # 统计列表的长度，即行数
        print(f"文件 {file_path} 中共有 {line_count} 行。")
        for line in lines:
            codeTokenNum.append(len(json.loads(line)['code_tokens']))
            nlTokenNum.append(len(json.loads(line)['docstring_tokens']))
except FileNotFoundError:
    print(f"找不到文件 {file_path}。")
except Exception as e:
    print("发生错误:", e)
```

```python
import pandas as pd
df = pd.DataFrame(codeTokenNum)
df.plot.hist(bins=50)
```

```python
df = pd.DataFrame(nlTokenNum)
df.plot.hist(bins=50)
```

```python
count = 0
for index in range(len(codeTokenNum)):
    if codeTokenNum[index]<100 and nlTokenNum[index]<20: count+=1
print(count)
```
