```python
import json
with open("data3/valid2.jsonl",'r') as file:
    index = 0
    for line in file:
        doc = json.loads(line)['docstring_tokens']
        for word in doc:
            if word == 'message':
                print(index,doc)
        index+=1
```

```python
from nltk.translate.bleu_score import sentence_bleu
with open("data3/bleu.txt",'r') as file:
    content = []
    for line in file:
        if line == '\n':
            if len(content) == 0:continue
            reference = [content[1]]
            candidate = content[0]
            bleu_score = sentence_bleu(reference, candidate)
            print(f'BLEU Score: {bleu_score:.4f}')
            content = []
            continue
        text = '{"test":'+line.rstrip().replace("'", "\"")+'}'
        doc = json.loads(text)
        content.append(doc['test'])
        print(doc['test'])
```
